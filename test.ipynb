calculating a weighted sum

```python
issueWarn=[]
for issue in sample_out:
    weighted_sum=0
    if issue['piiLeak']:
        if 'title' in issue:
            for leak in issue['title']:
                weighted_sum+=leak['score']*weights[leak['entity_group']]

        if 'description' in issue:
            for leak in issue['description']:
                weighted_sum+=leak['score']*weights[leak['entity_group']]

        if 'comments' in issue:
            for comment in issue['comments']:
                if comment["PII_leak"]:
                    for leak in comment["Leaks"]:
                      weighted_sum+=leak['score']*weights[leak['entity_group']]
        
        if 'attachments' in issue:
            for attachment in issue['attachments']:
                if comment["PII_leak"]:
                    for leak in comment["Leaks"]:
                      weighted_sum+=leak['score']*weights[leak['entity_group']]
        issueWarn.append({
            'id':issue['id'],
            'weighted_score':weighted_sum
        })

issueWarn
        
```

Thresholding

```python
for issue in issueWarn:
    if issue['weighted_score']< 8:
        print("low risk")
    elif issue['weighted_score']<15:
        print("mid risk")
    else:
        print("high risk")
```
